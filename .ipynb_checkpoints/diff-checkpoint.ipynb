```python
stop_90 = pd.read_csv('data/stop_90_euler.csv', sep=",")
stop_90_without_speed = stop_90[stop_90.columns[:-1]]
stop_90_without_speed
```

```python
euler=stop_90[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]
euler
```

```python
stop_90_diff = stop_90_without_speed.diff()
stop_90_diff
```

```python
stop_90_diff[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]=euler
stop_90_diff
```

```python
forward_90 = pd.read_csv('data/0_90_euler.csv', sep=",")
forward_90_without_speed = forward_90[forward_90.columns[:-1]]
euler_forward_90=forward_90[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]
forward_90_diff = forward_90_without_speed.diff()
forward_90_diff[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]=euler_forward_90
forward_90_diff
```

```python
left_90 = pd.read_csv('data/90_90_euler.csv', sep=",")
left_90_without_speed =left_90[left_90.columns[:-1]]
euler_left_90=left_90[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]
left_90_diff =left_90_without_speed.diff()
left_90_diff[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]=euler_left_90
left_90_diff
```

```python
backward_90 = pd.read_csv('data/180_90_euler.csv', sep=",")
backward_90_without_speed = backward_90[backward_90.columns[:-1]]
euler_backward_90=backward_90[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]
backward_90_diff = backward_90_without_speed.diff()
backward_90_diff[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]=euler_backward_90
backward_90_diff
```

```python
right_90 = pd.read_csv('data/270_90_euler.csv', sep=",")
right_90_without_speed = right_90[right_90.columns[:-1]]
euler_right_90=right_90[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]
right_90_diff = right_90_without_speed.diff()
right_90_diff[['e_r_x_euler','e_r_y_euler','e_r_z_euler', 'd_r_x_euler','d_r_y_euler','d_r_z_euler']]=euler_right_90
right_90_diff
```

```python
stop_90_diff.to_csv('stop_90_diff_euler.csv', index=False, header=True)
forward_90_diff.to_csv('0_90_diff_euler.csv', index=False, header=True)
left_90_diff.to_csv('90_90_diff_euler.csv', index=False, header=True)
backward_90_diff.to_csv('180_90_diff_euler.csv', index=False, header=True)
right_90_diff.to_csv('270_90_diff_euler.csv', index=False, header=True)
```
