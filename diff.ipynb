```python
from pathlib import Path
import numpy as np
import pandas as pd
```

```python
stop_90 = pd.read_csv('data/stop_90.csv', sep=",")
stop_90_without_speed = stop_90[stop_90.columns[:-1]]
stop_90_diff = stop_90_without_speed.diff()
```

```python
forward_90 = pd.read_csv('data/0_90.csv', sep=",")
forward_90_without_speed = forward_90[forward_90.columns[:-1]]
forward_90_diff = forward_90_without_speed.diff()
```

```python
left_90 = pd.read_csv('data/90_90.csv', sep=",")
left_90_without_speed = left_90[left_90.columns[:-1]]
left_90_diff =left_90_without_speed.diff()
```

```python
backward_90 = pd.read_csv('data/180_90.csv', sep=",")
backward_90_without_speed = backward_90[backward_90.columns[:-1]]
backward_90_diff =backward_90_without_speed.diff()
```

```python
right_90 = pd.read_csv('data/270_90.csv', sep=",")
right_90_without_speed = right_90[right_90.columns[:-1]]
right_90_diff = right_90_without_speed.diff()
```

```python
stop_90_diff.to_csv('stop_90_diff_without_axis.csv', index=False, header=True)
forward_90_diff.to_csv('0_90_diff_without_axis.csv', index=False, header=True)
left_90_diff.to_csv('90_90_diff_without_axis.csv', index=False, header=True)
backward_90_diff.to_csv('180_90_diff_without_axis.csv', index=False, header=True)
right_90_diff.to_csv('270_90_diff_without_axis.csv', index=False, header=True)
```
